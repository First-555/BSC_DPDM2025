### Binning

ใช้ histogram ดู bin ที่เหมาะสม

```python
plt.hist(flood['gfd_area'],30);
```

```python
outputhist = plt.hist(flood['gfd_area'],7)
```

```python
outputhist[1]
```

```python

def put_value_into_bins(value, bin_edges=outputhist[1]):
  for i in range(len(bin_edges)-1):
    if value >= bin_edges[i] and value < bin_edges[i+1]:
      new_value = (bin_edges[i] + bin_edges[i+1]-bin_edges[i])/2
      return new_value
  return None
```

```python
put_value_into_bins(3000)
```

```python
flood['gfd_area_bin']= flood['gfd_area'].apply(lambda x: put_value_into_bins(x,outputhist[1]))
```

```python
plt.plot(flood['gfd_area_bin'])
```

```python
plt.figure(figsize=(10, 8)) # Optional: Adjust figure size for better visibility
plt.subplot(2,1,1) # 2 rows, 1 column, first plot
plt.plot(flood['gfd_area'])
plt.title('Original gfd_area') # Add title for clarity
plt.subplot(2,1,2) # 2 rows, 1 column, second plot
plt.plot(flood['gfd_area_bin'])
plt.title('Binned gfd_area_bin') # Add title for clarity
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
